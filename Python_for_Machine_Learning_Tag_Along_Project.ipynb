```python
# Importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# A quick sneak peek into our dataset
df = pd.read_csv("FoodBalanceSheets_E_Africa_NOFLAG.csv", encoding = "latin-1")
df.head()
```

```python
# Conducting a summary statistic
df.describe()
```

```python
#5. What is the total sum of Wine produced in 2015 and 2018 respectively?
df.groupby('Item')[['Y2015', 'Y2018']].sum().loc['Wine']
```

```python
#7. Select columns Y2017 and Area, Perform a groupby operation on Area.  Which of these Areas had the highest sum in 2017?
df.groupby('Area').sum().sort_values(ascending=False, by='Y2017')[['Y2017']]
```

```python
#10. Perform a groupby operation on Element.  What year has the highest sum of Stock Variation?
df.groupby('Element').sum().loc[['Stock Variation']]
```

```python
#11. What is the mean and standard deviation across the whole dataset for the year 2017 to 2 decimal places?
df.describe()
```

```python
#14. Which year had the least correlation with Element Code?
df.corr().loc[['Element Code']]
```

```python
#15. Perform a groupby operation on Element.  What is the total number of the sum of Processing in 2017?
df.groupby('Element').sum().sort_values(ascending=False, by='Y2017')[['Y2017']].loc[['Processing']]
```

```python
#17. What is the total number of unique countries in the dataset?
df.Area.nunique()
```

```python
#18 Select columns Y2017 and Area, Perform a groupby operation on Area.  Which of these Areas had the 7th lowest sum in 2017?
df.groupby('Area').sum().sort_values(by='Y2017')[['Y2017']].iloc[6]
```

```python
#19. What is the total Protein supply quantity in Madagascar in 2015?
df.groupby(['Element', 'Area']).sum().loc['Protein supply quantity (g/capita/day)', 'Madagascar']
```

```python
#20. What is the total number and percentage of missing data in 2014 to 3 decimal places?
missing = df.Y2014.isnull().sum()
print(missing)
print(f"{missing/len(df):.3%}")
```
